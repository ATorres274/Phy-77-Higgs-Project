```python
from scipy.stats import norm
from scipy.optimize import curve_fit
%matplotlib inline

y, bin_edges = np.histogram(df11_df12_invMass, range=(100,160),bins=30)

#create an array of bin centers
x = 0.5*(bin_edges[1:] + bin_edges[:-1])
#avoid zero error, create a minimum error of 1
error_y = np.sqrt(y)
error_y = [max(error,1) for error in error_y]

#plot
plt.errorbar(x,y,error_y, fmt='o', label = 'Data')
plt.title('Distributions of the invariant mass')
plt.xlabel('mγγ GeV')
plt.ylabel('Events')
plt.legend()

print('Invariant Mass Data:\n')
print('Mean: %8.3f GeV'%np.mean(df11_df12_invMass))
print('STD: %9.3f'%np.std(df11_df12_invMass))
print()


#Curve fitting: define a 4th order polynomial
def model(x, a0, a1, a2, a3, a4):
    f = a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4
    return f

#From scipy.org & Workshop 7
popt, pcov = curve_fit(model, x, y)

#Display the useful data from this fit
print('Curve Fit Data:\n')

print('Optimal parameters: \n', popt, '\n')
r = y - model(x, *popt)
print('residuals: \n',r, '\n')
perr = np.sqrt(np.diag(pcov))
print('one standard deviation errors on the parameters: \n %.4f, %.4f, %.4f, %.4f, %.4f\n'% tuple(perr), '\n')
#chisq = sum((r / sigma) ** 2)


#plot the fit over data
plt.plot(x, model(x, *popt), 'r--', label= 'Bkg (4th order polynomial)')

#Curve Fitting with Signal+background:
mean = 126
sigma = (3.9 / (2*np.sqrt(2*np.log(2))))
A = 126.5





#define a 4th order polynomial plus signal
def model2(x, a0, a1, a2, a3, a4):
    return a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4 + 126.5*np.e**( (-(x-126)**2) / (2*(3.9 / (2*np.sqrt(2*np.log(2))))**2) )


par, cov = curve_fit(model2, x, y)

print('par: %.3f, %.3f, %.3f, %.3f, %.7f' % tuple(par))


#plot the fit over data
plt.plot(x, model2(x, *par), 'r-', label= 'Sig+Bkg (4th order polynomial)')
plt.legend()
plt.show()
```

```python
# Aaron's code for Z-score

import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt

'''
N = len(df11["invariant_mass"])

expected = N*np.ones(len(y))

en = np.ones(len(y))
chi2 = 0

for i in range(0, len(y)):
    value = y[i]
    en[i] = np.sqrt(expected[i])
    
    chi2 += ((value - expected[i])/en[i]) ** 2
    
ndf = len(y) - 1
print('chi2 = {chi2:4.2f} for {ndf:d} df'.format(chi2=chi2,ndf=ndf))

# compute chi and p
chi2_sp, p_sp = sp.stats.chisquare(f_obs=y,f_exp=expected)
print('chi2 from stats package = {0:4.2f}'.format(chi2_sp))
print('p-value from stats package = {0:4.2f}'.format(p_sp))
'''

# Chi2 for bg df11
y11, bin_edges11 = np.histogram(df11["invariant_mass"], range=(100,160),bins=30)
x11 = 0.5*(bin_edges[1:] + bin_edges[:-1])
popt11bg, pcov11bg = curve_fit(model, x11, y11)


chi2_bg11, p_bg11 = sp.stats.chisquare(f_obs = y11, f_exp = model(x11, *popt11bg))

print('chi2_5 (7 TeV) = {0:4.2f}'.format(chi2_bg11))
#print('p-value(?) = {0:4.2f}'.format(p_bg11))

# Chi2 for bg df12
y12, bin_edges12 = np.histogram(df12["invariant_mass"], range=(100,160),bins=30)
x12 = 0.5*(bin_edges[1:] + bin_edges[:-1])
popt12bg, pcov12bg = curve_fit(model, x12, y12)

chi2_bg12, p_bg12 = sp.stats.chisquare(f_obs = y12, f_exp = model(x12, *popt12bg))

print('chi2_5 (8 TeV) = {0:4.2f}'.format(chi2_bg12))
#print('p-value(?) = {0:4.2f}'.format(p_bg12))


# Chi2 for sig df11 
popt11sig, pcov11sig = curve_fit(model2, x11, y11)
chi2_sig11, p_sig11 = sp.stats.chisquare(f_obs = y11, f_exp = model2(x11, *popt11sig))

print('chi2_6 (7 TeV) = {0:4.2f}'.format(chi2_sig11))
#print('p-value(?) = {0:4.2f}'.format(p_sig11))

# Chi2 for sig df12
popt12sig, pcov12sig = curve_fit(model2, x12, y12)
chi2_sig12, p_sig12 = sp.stats.chisquare(f_obs = y12, f_exp = model2(x12, *popt12sig))

print('chi2_6 (8 TeV) = {0:4.2f}\n'.format(chi2_sig12))
#print('p-value(?) = {0:4.2f}'.format(p_sig12))

# Change in chi2
dchi2_11 = chi2_bg11 - chi2_sig11
dchi2_12 = chi2_bg12 - chi2_sig12

dchi2_tot = dchi2_11 + dchi2_12

print('delta chi2 (7 TeV) = {0:4.2f}'.format(dchi2_11))
print('delta chi2 (8 TeV) = {0:4.2f}'.format(dchi2_12))
print('delta chi2 total = {0:4.2f}\n'.format(dchi2_tot))


# Z-scores
z_score11 = np.sqrt(dchi2_11)
z_score12 = np.sqrt(dchi2_12)

z_score_tot = np.sqrt(z_score11**2 + z_score12**2)

print('Z-Score (7 TeV) = {0:4.2f}'.format(z_score11))
print('Z-Score (8 TeV) = {0:4.2f}'.format(z_score12))
print('Z-Score Total = {0:4.2f}\n'.format(z_score_tot))


# Manual Chi2 and Z-score
# Error 11
error_y11 = np.sqrt(y11)
error_y11 = [max(error,1) for error in error_y11]

# Error 12
error_y12 = np.sqrt(y12)
error_y12 = [max(error,1) for error in error_y12]

# Chi2 bg 11
chi_squared_bg11 = np.sum(((model(x11, *popt11bg) - y11) / error_y11) ** 2)
#chi_squared_bg11 = np.sum((y11 - model(x11, *popt11bg)) ** 2 / model(x11, *popt11bg))
print('manual chi2_5 (7 TeV) = {0:4.2f}'.format(chi_squared_bg11))

# Chi2 sig 11
chi_squared_sig11 = np.sum(((model2(x11, *popt11sig) - y11) / error_y11) ** 2)
#chi_squared_sig11 = np.sum((y11 - model2(x11, *popt11sig)) ** 2 / model2(x11, *popt11sig))
print('manual chi2_6 (7 TeV) = {0:4.2f}'.format(chi_squared_sig11))

# Chi2 bg 12
chi_squared_bg12 = np.sum(((model(x12, *popt12bg) - y12) / error_y12) ** 2)
print('manual chi2_5 (8 TeV) = {0:4.2f}'.format(chi_squared_bg12))

# Chi2 sig 12
chi_squared_sig12 = np.sum(((model2(x12, *popt12sig) - y12) / error_y12) ** 2)
print('manual chi2_6 (8 TeV) = {0:4.2f}\n'.format(chi_squared_sig12))

# Delta Cchi2
dchi_squared_11 = chi_squared_bg11 - chi_squared_sig11
dchi_squared_12 = chi_squared_bg12 - chi_squared_sig12
dchi_squared_tot = dchi_squared_11 + dchi_squared_12

print('manual delta chi2 (7 TeV) = {0:4.2f}'.format(dchi_squared_11))
print('manual delta chi2 (8 TeV) = {0:4.2f}'.format(dchi_squared_12))
print('manual delta chi2 total = {0:4.2f}\n'.format(dchi_squared_tot))


# Z-Score
mz_score11 = np.sqrt(dchi_squared_11)
mz_score12 = np.sqrt(dchi_squared_12)
mz_score_tot = np.sqrt(mz_score11**2 + mz_score12**2)

print('manual Z-Score (7 TeV) = {0:4.2f}'.format(mz_score11))
print('manual Z-Score (8 TeV) = {0:4.2f}'.format(mz_score12))
print('manual Z-Score Total = {0:4.2f}\n'.format(mz_score_tot))
```
