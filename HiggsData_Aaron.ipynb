```python
import uproot as ur
file11 = ur.open("http://moller.physics.berkeley.edu/~yury/Phys129/tree_TYPE_DATA_2011.root")
file12 = ur.open("http://moller.physics.berkeley.edu/~yury/Phys129/tree_TYPE_DATA_2012.root")
print(file11.keys())
print(file12.keys())
tree11 = file11["TYPE_DATA"]
tree12 = file12["tree_TYPE_DATA"]
print(tree11.keys())
print(tree12.keys())
```

```python
import matplotlib.pyplot as plt
import numpy as np

#df = tree.pandas.df(["invariant_mass","ph*","categ*"])
#df12 = tree12.pandas.df()
#df12
df11 = tree11.pandas.df()
df11

df12 = tree12.pandas.df()
df12
```

```python
plt.yscale('linear')
df12["invariant_mass"].plot.hist(bins=30)
plt.show()

plt.yscale('linear')
df11["invariant_mass"].plot.hist(bins=30)
plt.show()
```

```python
mean11 = np.mean(df11["invariant_mass"])
mean12 = np.mean(df12["invariant_mass"])

std11 = np.std(df11["invariant_mass"])
std12 = np.std(df12["invariant_mass"])

print("Mean df11: %.3f" %mean11)
print("Mean df12: %.3f" %mean12)

print("STD df11: %.3f" %std11)
print("STD df12: %.3f" %std12)

z_scores11 = (df11["invariant_mass"] - mean11) / std11
z_scores12 = (df12["invariant_mass"] - mean12) / std12

print(z_scores11)
print(z_scores12)
```
