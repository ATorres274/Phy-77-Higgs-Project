```python
##### Curve Fitting Models Unweighted #####
# a0 is the number of events in the histogram at m = 130
# a0*b1 is the rough slope of the graph
# b1, b2, b3, b4 should be fairly small

def bgmodel(m, a0, b1, b2, b3, b4):
    return a0*(1 + b1*(m-130) + b2*(m-130)**2 + b3*(m-130)**3 + b4*(m-130)**4)

def sigmodel(m, a0, b1, b2, b3, b4, A):
    return a0*(1 + b1*(m-130) + b2*(m-130)**2 + b3*(m-130)**3 + b4*(m-130)**4) + A*(np.exp(-0.5*((m-mu)/sigma)**2)/(sigma*np.sqrt(2*np.pi)))

##### Plotting Distribution Models #####

def dist_bgmodel(x, a0, a1, a2, a3, a4):
    f = a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4
    return f

#define a 4th order polynomial plus signal
def dist_sigmodel(x, a0, a1, a2, a3, a4):
    return a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4 + 126.5*np.e**( (-(x-126)**2) / (2*(3.9 / (2*np.sqrt(2*np.log(2))))**2) )
```

```python
##### Functions to create Category Plots #####
def Cat10_Data_Plotting(cat10, nbins, par0_bg, a0_sig):
    ## Subplot creation and labeling ##
    fig, ax = plt.subplots(5, 2, figsize=(30, 30), sharex='col', gridspec_kw={'hspace': 0})
    ax[4, 0].set_xlabel(r'$m_{\gamma \gamma}$ [GeV]')
    ax[4, 1].set_xlabel(r'$m_{\gamma \gamma}$ [GeV]')

    # histogram outputs
    n = []
    bins = []

    k = 0
    # Main plotting loop
    for i in range(2):
        for j in range(5):
            n.append([])
            bins.append([])
    
            n[k], bins[k], _ = ax[j, i].hist(cat10[k], color='r', alpha=0.1, bins=nbins)
            ax[j, i].set_ylabel('Invariant Mass: Category %i' % (k + 1))
            ax[j, i].set_xlim(100, 160)
        
            k += 1

    ## Error Calculation ##
    err = np.sqrt(n)
    
    for i in range(10):
        err[i] = [max(error, 1) for error in err[i]]
        
    # x data for fitting
    x = []
    
    for i in range(10):
        x.append(0.5*(bins[i][1:] + bins[i][:-1]))
    
    # Fitting
    # Background
    opt_bg = []
    cov_bg = []

    # Background + Signal
    opt_sig = []
    cov_sig = []

    par0_sig = [] # filled in loop after finding opt7_bg
    for i in range(10):
        # Background curve fitting
        opt_bg.append([])
        cov_bg.append([])
        opt_bg[i], cov_bg[i]= curve_fit(bgmodel, xdata=x[i], ydata=n[i], p0=par0_bg[i], sigma=err[i])
    
        # Background + Signal curve fitting
        par0_sig.append([*opt_bg[i], a0_sig[i]])
        opt_sig.append([])
        cov_sig.append([])
        opt_sig[i], cov_sig[i] = curve_fit(sigmodel, xdata=x[i], ydata=n[i], p0=par0_sig[i], sigma=err[i])

    ## Curve plotting ##
    # x-values for plotting
    massRng = np.linspace(100, 160, 1000)

    k = 0
    for i in range(2):
        for j in range(5):
            # Background
            ax[j, i].plot(massRng, bgmodel(massRng, *opt_bg[k]), color='r', linestyle=':', linewidth=2.5, label='Bkg (4th-Order Polynomial)')
        
            # Background + Signal
            # Why are we using the par0s not opts?
            ax[j, i].plot(massRng, sigmodel(massRng, *par0_sig[k]), color='r', linestyle='-', linewidth=2.5, label='Sig + Bkg Fit')
            k += 1

    ## Error bar Plotting ##
    k = 0
    for i in range(2):
        for j in range(5):
            ax[j, i].errorbar(x[k], n[k], yerr=err[k], color='k', fmt='o', label='Data')
            k += 1
            
    ## Plot Wrap ##
    for i in range(2):
        for j in range(5):
            ax[j, i].legend(fontsize='xx-large')
    fig.tight_layout()
    plt.show()
    return
```

```python
##### Functions to create "Total" Plots #####
def Total_Data_Plotting(mass_data, par0Tot_bg, a0Tot_sig):
    ## Subplot Creation and Labeling ##
    figTot, axTot = plt.subplots(4, figsize=(20, 15), sharex='col', gridspec_kw={'hspace': 0})

    axTot[0].remove()
    axTot[1].remove()
    axTot[2].remove()

    axTot[0] = figTot.add_subplot(axTot[0].get_gridspec()[0:3])
    nTot, binTot, _ = axTot[0].hist(mass_data, color='r', alpha=0.1, bins=30)
    xTot = 0.5*(binTot[1:] + binTot[:-1])

    axTot[0].set_ylabel('Events')
    axTot[3].set_ylabel('Events - Bkg')
    axTot[3].set_xlabel(r'$m_{\gamma \gamma}$ [GeV]')

    axTot[0].set_xticks([])
    axTot[0].set_xlim(100, 160)
    axTot[3].set_xlim(100, 160)
    axTot[3].set_ylim(-200, 200)
    
    ## Computing Errors ##   
    errTot = np.sqrt(nTot)
    errTot  = [max(error, 1) for error in errTot]
    
    ## Curve Fitting ##
    optTot_bg, covTot_bg = curve_fit(bgmodel, xdata=xTot, ydata=nTot, p0=par0Tot_bg, sigma=err7Tot)
    optTot_sig, covTot_sig = curve_fit(sigmodel, xdata=xTot, ydata=nTot, p0=a0Tot_sig, sigma=err7Tot)

    ## Curve Plotting ##
    # Summed Weights Distribution and Difference
    axTot[0].plot(massRng, bgmodel(massRng, *optTot_bg), color='r', linestyle=':', linewidth=2.5, label='Bkg (4th-Order Polynomial)')
    axTot[3].plot(massRng, bgmodel(massRng, *optTot_bg) - bgmodel(massRng, *optTot_bg), color='r', linestyle=':', linewidth=2.5)

    axTot[0].plot(massRng, sigmodel(massRng, *a0Tot_sig), color='r', linestyle='-', linewidth=2.5, label='Sig + Bkg Fit')
    axTot[3].plot(massRng, sigmodel(massRng, *a0Tot_sig) - bgmodel(massRng, *optTot_bg), color='r', linestyle='-', linewidth=2.5)

    ## Error Bar Plotting ##
    axTot[0].errorbar(xTot, nTot, yerr=errTot, color='k', fmt='o', label='Data')
    axTot[3].errorbar(xTot, nTot - bgmodel(xTot, *optTot_bg), yerr=errTot, color='k', fmt='o')

    ## Legends ##
    axTot[0].legend()
    figTot.tight_layout()

    plt.show()

    return
```

```python
def Plot_Dist_Inv_Mass(mass_data):
    ## Plots Histogram
    y, bin_edges = np.histogram(mass_data, range=(100,160),bins=30)
    
    #create an array of bin centers
    x = 0.5*(bin_edges[1:] + bin_edges[:-1])

    #avoid zero error, create a minimum error of 1
    error_y = np.sqrt(y)
    error_y = [max(error,1) for error in error_y]

    #plot
    plt.errorbar(x,y,error_y, fmt='o', label = 'Data')
    plt.title('Distributions of the invariant mass')
    plt.xlabel('mγγ GeV')
    plt.ylabel('Events')
    plt.legend()

    ## fit a curve to the Bkg function ##
    #From scipy.org & Workshop 7
    popt, pcov = curve_fit(dist_bgmodel, x, y)
    
    #Display the useful data from this fit
    print('Background Curve Fit Data:\n')

    #Save the coefficients of the polynomial
    a_bg = []
    error_bg = []
    print('Values that maximize 𝜒2 for Bkg:')
    
    for i in range(len(popt)):
        a_bg.append(popt[i])
        error_bg.append(np.sqrt(pcov[i,i]))
        print('a%i= %.3f ± %.2f' % (i,a_bg[i] , error_bg[i])       )
    
    print()

    chi_squared_bkg = np.sum(((dist_bgmodel(x, *popt)-y)/sigma)**2)
    reduced_chi_squared_bkg = (chi_squared_bkg)/(len(x)-len(popt))
    
    print('Bkg fit 𝜒2 = %.2f' % chi_squared_bkg)
    print ('Reduced 𝜒2 = %f' % reduced_chi_squared_bkg, '\n')

    #plot the fit over data
    plt.plot(x, dist_bgmodel(x, *popt), 'r--', label= 'Bkg (4th order polynomial)')

    print()

    #fit a curve to the Sig+Bkg function
    par, cov = curve_fit(dist_sigmodel, x, y)
    
    #Curve Fitting with Signal+background:
    print('Signal + Background Curve Fit Data:\n')

    
    #Save the coefficients of the polynomial
    a_sig = []
    error_sig = [] 
    print('Values that maximize 𝜒2 for Sig+Bkg:')
    
    for i in range(len(par)):
        a_sig.append(par[i])
        error_sig.append(np.sqrt(pcov[i,i]))
        print('a%.i= %.3f ± %.2f' % (i, a_sig[i] , error_sig[i])       )
        
    print()

    chi_squared_sigBkg = np.sum(((dist_sigmodel(x, *par)-y)/sigma)**2)
    reduced_chi_squared_sigBkg = (chi_squared_sigBkg)/(len(x)-len(par))

    print('Sig+Bkg fit 𝜒2 = %.2f' % chi_squared_sigBkg)
    print ('Reduced 𝜒2 = %f' % reduced_chi_squared_sigBkg, '\n')

    #plot the fit over data
    plt.plot(x, dist_sigmodel(x, *par), 'r-', label= 'Sig+Bkg (4th order polynomial)')
    plt.legend()
    plt.show()
    
    return
```
