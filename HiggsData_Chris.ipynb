```python
#Combine df11 & df12
df11_df12_invMass = np.concatenate((df11["invariant_mass"], df12["invariant_mass"]))
```

```python
df11["invariant_mass"].plot.hist(bins=30)
```

```python
from scipy.stats import norm
from scipy.optimize import curve_fit
%matplotlib inline

y, bin_edges = np.histogram(df11_df12_invMass, range=(100,160),bins=30)

#create an array of bin centers
x = 0.5*(bin_edges[1:] + bin_edges[:-1])

#avoid zero error, create a minimum error of 1
error_y = np.sqrt(y)
error_y = [max(error,1) for error in error_y]

#plot
plt.errorbar(x,y,error_y, fmt='o', label = 'Data')
plt.title('Distributions of the invariant mass')
plt.xlabel('mγγ GeV')
plt.ylabel('Events')
plt.legend()

print('Invariant Mass Data:')
print('Mean: %8.3f GeV'%np.mean(df11_df12_invMass))
print('STD: %9.3f'%np.std(df11_df12_invMass))
print()



#Curve fitting: define a 4th order polynomial
def model(x, a0, a1, a2, a3, a4):
    return a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4

#From scipy.org & Workshop 7
popt, pcov = curve_fit(model, x, y)

print('Curve Fit Data:')
perr = np.sqrt(np.diag(pcov))
print('Sigma on the parameters: \t %.4f, %.4f, %.4f, %.4f, %.4f\n'% tuple(perr))

#plot the fit over data
plt.plot(x, model(x, *popt), 'r--', label= 'Bkg (4th order polynomial)')
plt.legend()


#Curve Fitting with background:
```
